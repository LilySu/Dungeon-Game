```
import re
path = '/content/drive/My Drive/Colab Notebooks/dungeon_name.txt.txt'

with open(path, 'r',encoding='cp1252') as f:
    x = f.read().splitlines()

stopwords = ['ve',"n't",'check-in','=','= =', ' ', 'it', "it's", 'it.','they', 'the', 'this','its', 'l','-','they','this',"don't",'the ', ' the', 'it', 'i"ve', 'i"m', '!', '1','2','3','4', '5','6','7','8','9','0','/','.',',','?']

def filter_stopwords(text):
  nonstopwords = []
  for i in text:
    if i not in stopwords:
      nonstopwords.append(i)
  return nonstopwords

x = [i for i in x if re.search(r'[a-zA-Z]', i)]

x = [i.split(' ') for i in x]

x = [item for sublist in x for item in sublist]

x = [i.replace('.','') for i in x]
x = [i.replace('?','') for i in x]
x = [i.replace('—','') for i in x]

# x
```

```
# Dungeon World by Jonathan Brooks 

# import urllib.request
# import random

# word_url = "http://svnweb.freebsd.org/csrg/share/dict/words?view=co&content-type=text/plain"
# response = urllib.request.urlopen(word_url)
# long_txt = response.read().decode()
# words = long_txt.splitlines()

# upper_words = [word for word in words if word[0].isupper()]
# name_words  = [word for word in upper_words if not word.isupper()]
# one_name = ' '.join([name_words[random.randint(0, len(name_words))] for i in range(2)])

def find_adj(x):
  """Get Just the Adjectives"""
  adj_list = []
  doc = nlp(str(x))
  for token in doc:
    if token.pos_ == 'ADJ':
      adj_list.append(token.orth_)
  return adj_list

adj_list = find_adj(x)

def find_noun(x):
  """Get Just the Adjectives"""
  noun_list = []
  doc = nlp(str(x))
  for token in doc:
    if token.pos_ == 'NOUN':
      noun_list.append(token.orth_)
  return noun_list

noun_list = find_noun(x)
```
